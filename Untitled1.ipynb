```python
# Quantum Radar Sistemi (önceki kodundan)
class QuantumRadarSystem:
    def __init__(self, num_photons=8000, correlation=0.96):
        self.num_photons = num_photons
        self.correlation = correlation
    
    def quantum_source(self):
        noise = np.random.normal(0, 0.1, self.num_photons)
        signal = np.random.normal(0, 1, self.num_photons) + noise
        idler = self.correlation * signal + np.random.normal(0, 0.3, self.num_photons)
        print(f"Quantum Source: {self.num_photons} dolanık foton çifti üretildi.")
        return signal, idler
    
    def transmitter(self, signal, target_rcs=0.08, distance=1500):
        reflection = signal * np.sqrt(target_rcs)
        received = reflection + np.random.normal(0, 1.1, self.num_photons)
        print(f"Transmitter: RCS={target_rcs} | Menzil={distance} m")
        return received
    
    def signal_processor(self, idler, received, threshold=0.015):
        print("\n=== Signal Processor ===")
        phase = np.angle(np.correlate(idler, received, mode='valid'))
        corrected = received * np.exp(-1j * phase)
        filtered = corrected * np.conj(idler) / (np.abs(idler)**0.5 + 1e-6)
        
        mean_f = np.mean(np.abs(filtered))
        std_f = np.std(np.abs(filtered))
        snr = mean_f / std_f if std_f > 0 else 0
        
        final_corr = np.corrcoef(np.abs(idler), np.abs(filtered))[0,1]
        detection_score = (final_corr ** 2) * (snr ** 1.5)
        
        detected = detection_score > threshold
        print(f"Final Correlation: {final_corr:.4f}")
        print(f"SNR: {snr:.3f}")
        print(f"Tespit Skoru: {detection_score:.4f}")
        print(f"Hedef Tespit Edildi mi? → {'EVET' if detected else 'HAYIR'}")
        return final_corr, detection_score, detected

# Klasik Monostatik Radar
class MonostaticRadar:
    def __init__(self):
        self.c = 3e8
        self.freq = 10e9
        self.lambda_ = self.c / self.freq
        self.pt = 1000
        self.g = 100
    
    def radar_equation(self, rcs, distance):
        pr = (self.pt * self.g**2 * self.lambda_**2 * rcs) / ((4 * np.pi)**3 * distance**4)
        noise = 1e-12
        snr = 10 * np.log10(pr / noise)
        print(f"Klasik Radar - RCS: {rcs} | Menzil: {distance} m | SNR: {snr:.2f} dB")
        return snr, pr
    
    def detect(self, rcs=0.08, distance=1500, threshold=12):
        snr, _ = self.radar_equation(rcs, distance)
        detected = snr > threshold
        print(f"  Tespit: {'EVET' if detected else 'HAYIR'}")
        return detected, snr

print("Sınıflar tanımlandı.")
```
